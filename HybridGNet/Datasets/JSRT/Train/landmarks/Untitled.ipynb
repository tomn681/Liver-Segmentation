```python
import numpy as np
import matplotlib.pyplot as plt

landmarks = np.load('./JPCLN002.npy')
landmarks = landmarks.astype('float').reshape(-1, 2)

mask = np.zeros((1024,1024,3))

for idx, element in enumerate(landmarks):
    x, y = element
    x, y = int(x), int(y)
    if idx < 44+50:
        ch = 0
    elif 44+50 < idx < 44+50+26:
        ch = 1
    else:
        ch = 2
    mask[y-5:y+5, x-5:x+5, ch] = 255
    
plt.imshow(mask)
```
