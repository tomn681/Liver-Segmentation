```python
from models.hybrid import Hybrid as Hybrid
from models.hybridSkip import Hybrid as Skip
from models.hybridDoubleSkip import Hybrid as DoubleSkip

import os 

import torch
from torchvision import transforms
import numpy as np

import pathlib
import re

from utils.dataLoader import LandmarksDataset, ToTensor, Rescale
from sklearn.metrics import mean_squared_error
import scipy.sparse as sp
from utils.utils import scipy_to_torch_sparse, genMatrixesLH

def natural_key(string_):
    """See http://www.codinghorror.com/blog/archives/001018.html"""
    return [int(s) if s.isdigit() else s for s in re.split(r'(\d+)', string_)]

device = 'cuda:0'
```

```python
abnormal = "../Datasets/Shenzhen_Abnormal/Prep/Images/"

data_root = pathlib.Path(abnormal)
all_files = list(data_root.glob('*.png'))
all_files = [str(path) for path in all_files]
all_files.sort(key = natural_key)

len(all_files)
```

```python
A, AD, D, U = genMatrixesLH()

A = sp.csc_matrix(A).tocoo()
AD = sp.csc_matrix(AD).tocoo()
D = sp.csc_matrix(D).tocoo()
U = sp.csc_matrix(U).tocoo()

D_ = [D.copy()]
U_ = [U.copy()]
A_ = [A.copy(), A.copy(), A.copy(), AD.copy(), AD.copy(), AD.copy()]

config = {}
config['n_nodes'] = [120, 120, 120, 60, 60, 60]

A_t, D_t, U_t = ([scipy_to_torch_sparse(x).to(device) for x in X] for X in (A_, D_, U_))

config['latents'] = 64
config['inputsize'] = 1024

f = 32
config['filters'] = [2, f, f, f, f//2, f//2, f//2]
config['skip_features'] = f

config['window'] = (3,3)

config['K'] = 6
config['l1'] = 6
config['l2'] = 5

double65 = DoubleSkip(config.copy(), D_t, U_t, A_t).to(device)
double65.load_state_dict(torch.load("../weights/Skip/double_L65/best.pt"))
double65.eval()
print('Model loaded')
```

```python
from models.unet import UNet

unet = UNet(n_classes = 3).to(device)
unet.load_state_dict(torch.load('../weights/UNet/best.pt'))
unet.eval()

print('Model loaded')
```

```python
from medpy.metric import dc, hd, assd, jc
from utils.fun import reverseVector, drawBinary

def evalImageMetrics(blank, output, target):
    p1, p2, h, c1, c2 = reverseVector(output)
    
    ptest = drawBinary(blank.copy(), p1)
    ptest = drawBinary(ptest, p2)

    hdp = hd(ptest, target == 1)
    dcp = dc(ptest, target == 1)
    
    return [dcp, hdp]

def evalImageMetricsUNet(output, target):
    dcp = dc(output == 1, target == 1)
    
    hdp = hd(output == 1, target == 1)
    
    return [dcp, hdp]
```

```python
from utils.fun import drawOrgans
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

results = pd.DataFrame()
blank = np.zeros([1024, 1024])

i = 0

for sample in all_files:
    with torch.no_grad():
        img = cv2.imread(sample, 0) / 255
        data = np.expand_dims(img, axis = 0)
        data = np.expand_dims(data, axis = 0)
        data = torch.tensor(data).to(device).float()

        anatomical = cv2.imread(sample.replace('Images','Masks'), cv2.IMREAD_UNCHANGED) // 255
        air = cv2.imread(sample.replace('Images','Air'), cv2.IMREAD_UNCHANGED) // 255
        
        output = double65(data)
        if len(output) > 1:
            output = output[0]
            
        output = output.cpu().numpy().reshape(-1) 
        output = np.clip(output, 0, 1)[:240]
        
        metrics_air = evalImageMetrics(blank, output * 1024, air)
        metrics_anato = evalImageMetrics(blank, output * 1024, anatomical)
        
        aux = pd.DataFrame([['HybridGNet 2-IGSC', 'Air'] + metrics_air], columns=['Model', 'Annotation', 'Dice Lungs', 'HD Lungs'])
        results = results.append(aux, ignore_index = True)
        
        aux = pd.DataFrame([['HybridGNet 2-IGSC', 'Anatomy'] + metrics_anato], columns=['Model', 'Annotation', 'Dice Lungs', 'HD Lungs'])
        results = results.append(aux, ignore_index = True)
        
        output = unet(data)
        output = torch.argmax(output[0,:,:,:], axis=0).cpu().numpy()
        
        metrics_air = evalImageMetricsUNet(output, air)
        metrics_anato = evalImageMetricsUNet(output, anatomical)
        
        aux = pd.DataFrame([['UNet', 'Air'] + metrics_air], columns=['Model', 'Annotation', 'Dice Lungs', 'HD Lungs'])
        results = results.append(aux, ignore_index = True)
        
        aux = pd.DataFrame([['UNet', 'Anatomy'] + metrics_anato], columns=['Model', 'Annotation', 'Dice Lungs', 'HD Lungs'])
        results = results.append(aux, ignore_index = True)
```

```python
import seaborn as sns

plt.figure(figsize = (9,5))
plt.tight_layout()
ax = plt.subplot(1,2,1)
sns.boxplot(x = 'Model', y = 'Dice Lungs', data = results, hue ='Annotation', showmeans = True, palette = sns.color_palette("Pastel2"))
plt.xticks(rotation=25, ha="right" )
plt.ylabel('Dice')
plt.title('Dice Lungs')
plt.xlabel(None)

ax = plt.subplot(1,2,2)
sns.boxplot(x = 'Model', y = 'HD Lungs', data = results, hue ='Annotation', showmeans = True, palette = sns.color_palette("Pastel2"))
plt.xticks(rotation=25, ha="right" )
plt.ylabel('HD (pixels)')
plt.title('Hausdorff Distance')
plt.xlabel(None)

plt.tight_layout()
plt.savefig('figs/air_vs_anatomical.png', dpi=200, bbox_inches="tight")
plt.savefig('figs/air_vs_anatomical.pdf', dpi=200, bbox_inches="tight")
plt.show()
```

```python
def gt_color(draw, gt):
    image = np.zeros(list(draw.shape) + [3])
    image[:,:,0] = draw + 0.7 * (gt == 1).astype('float')
    image[:,:,1] = draw - 0.2 * (gt == 1).astype('float')
    image[:,:,2] = draw - 0.2 * (gt == 1).astype('float')
    image = np.clip(image, 0, 1)
    
    return image
```

```python
results = pd.DataFrame()
blank = np.zeros([1024, 1024])

i = [0,3,8]

plt.figure(figsize = (9,6), dpi = 200)

for j in range(0,3):
    with torch.no_grad():
        sample = all_files[i[j]]
        img = cv2.imread(sample, 0) / 255
        data = np.expand_dims(img, axis = 0)
        data = np.expand_dims(data, axis = 0)
        data = torch.tensor(data).to(device).float()

        anatomical = cv2.imread(sample.replace('Images','Masks'), cv2.IMREAD_UNCHANGED) // 255
        air = cv2.imread(sample.replace('Images','Air'), cv2.IMREAD_UNCHANGED) // 255
        
        anato_img = gt_color(img, anatomical)
        air_img = gt_color(img, air)
                
        plt.subplot(3,5,1 + j*5)
        if j==0:
            plt.title('Image', fontsize=10)
        plt.imshow(img, cmap='gray')
        plt.xlim(1, 1024)
        plt.ylim(1024, 1)
        plt.axis('off')
        
        plt.subplot(3,5,2+ j*5)
        if j==0:
            plt.title('Air mask', fontsize=10)
        plt.imshow(air_img)
        plt.xlim(1, 1024)
        plt.ylim(1024, 1)
        plt.axis('off')
        
        plt.subplot(3,5,3+ j*5)
        if j==0:
            plt.title('Anatomy mask', fontsize=10)
        plt.imshow(anato_img)
        plt.xlim(1, 1024)
        plt.ylim(1024, 1)
        plt.axis('off')
        
        output = double65(data)
        if len(output) > 1:
            output = output[0]
            
        output = output.cpu().numpy().reshape(-1) 
        output = np.clip(output, 0, 1)[:240]
        
        p1, p2, h, c1, c2 = reverseVector(output * 1024)
        lungs = drawBinary(blank.copy(), p1) 
        lungs = drawBinary(lungs, p2) / 255
        heart = drawBinary(blank.copy(), h) / 255
        output = lungs + 2 * heart
        output = np.clip(output, 0, 2)
        
        hybrid_seg = gt_color(img, output)
        
        plt.subplot(3,5,4+ j*5)
        if j==0:
            plt.title('HybridGNet 2-IGSC', fontsize=10)
        plt.imshow(hybrid_seg)
        plt.xlim(1, 1024)
        plt.ylim(1024, 1)
        plt.axis('off')
        
        output = unet(data)
        output = torch.argmax(output[0,:,:,:], axis=0).cpu().numpy()
        
        unet_seg = gt_color(img, output)
        
        plt.subplot(3,5,5+ j*5)
        if j==0:
            plt.title('UNet', fontsize=10)
        plt.imshow(unet_seg)
        plt.xlim(1, 1024)
        plt.ylim(1024, 1)
        plt.axis('off')
        
        plt.subplots_adjust(wspace=0, hspace=0)
        plt.savefig('figs/air_anato_masks.png', bbox_inches= 'tight', dpi = 200)
        plt.savefig('figs/air_anato_masks.pdf', bbox_inches= 'tight', dpi = 200)
```
